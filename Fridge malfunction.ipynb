```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor,ExtraTreeRegressor
from sklearn.ensemble import AdaBoostRegressor,BaggingRegressor,GradientBoostingRegressor,RandomForestRegressor,StackingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
models = [LGBMRegressor,AdaBoostRegressor,BaggingRegressor,GradientBoostingRegressor,XGBRegressor,RandomForestRegressor,DecisionTreeRegressor,ExtraTreeRegressor,LinearRegression,Lasso,Ridge,ElasticNet]
```

```python
names = ["LGBMRegressor","AdaBoostRegressor","BaggingRegressor","GradientBoostingRegressor","XGBRegressor","RandomForestRegressor","DecisionTreeRegressor","ExtraTreeRegressor","LinearRegression","Lasso","Ridge","ElasticNet"]
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
for i in range(0,len(models)):
    mod = models[i]()
    mod.fit(x_train,y_train)
    print(f"{names[i]} MSE score is {mean_squared_error(y_test,mod.predict(x_test))}")
    print(f"{names[i]} MAE score is {mean_absolute_error(y_test,mod.predict(x_test))}")
```

```python
nam = []
mse = []
mae = []
for i in range(0,len(models)):
    mod = models[i]()
    mod.fit(x_train,y_train)
    nam.append(names[i])
    mse.append(mean_squared_error(y_test,mod.predict(x_test)))
    mae.append(mean_absolute_error(y_test,mod.predict(x_test)))
    
```
